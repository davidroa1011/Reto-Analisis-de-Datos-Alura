```python
df_masVendidosTiendas = df_masVendidosTiendas.reset_index(drop=True)

df_pivot = df_masVendidosTiendas.pivot_table(
    index='Categoria del producto',
    columns='Tienda',
    values='Cantidad de ventas'
)

ax = df_pivot.plot(
    kind='line',
    figsize=(12, 6),
    marker='o'
)

plt.title('Comparación de Cantidad de Ventas por Categoría (Todas las Tiendas)', fontsize=16)
plt.xlabel('Categoría del Producto', fontsize=12) # Puedes usar la ortografía correcta para la ETIQUETA
plt.ylabel('Cantidad de Ventas', fontsize=12)
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.savefig("Cantidad de ventas por categoria por tienda.png", dpi = 300)
plt.show()
```
